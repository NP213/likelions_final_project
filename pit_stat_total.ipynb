```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv('data/KBO_2020_2025_투수_전체통합.csv')

# 2. 통계 계산 함수 정의
def get_team_stats(group):
    # (조건: IP >= 30) 데이터 필터링
    cond_group = group[group['IP'] >= 30]
    
    stats = {
        # 전체 선수 대상
        'IP': round(group['IP'].sum(), 3),
        'GS': group['GS'].sum(),
        'GR': group['GR'].sum(),
        'FIP': round(group['FIP'].mean(), 3),
        'FIP(중앙값)': round(group['FIP'].median(), 3),
        'WAR': round(group['WAR'].sum(), 3),
        
        # 조건부(IP >= 30) 선수 대상
        '(조건 IP>=30)IP': round(cond_group['IP'].sum(), 3),
        '(조건 IP>=30)GS': cond_group['GS'].sum(),
        '(조건 IP>=30)GR': cond_group['GR'].sum(),
        '(조건 IP>=30)FIP': round(cond_group['FIP'].mean(), 3) if not cond_group.empty else 0,
        '(조건 IP>=30)FIP(중앙값)': round(cond_group['FIP'].median(), 3) if not cond_group.empty else 0,
        '(조건 IP>=30)WAR': round(cond_group['WAR'].sum(), 3) 
    }
    return pd.Series(stats)

# 3. 팀(Team)과 연도(Year)로 그룹화하여 통계 산출
# 팀별로 모아서 보기 위해 Team을 우선 순위로 그룹화합니다.
final_result = df.groupby(['Team', 'Year']).apply(get_team_stats).reset_index()

# 4. 데이터 정렬 (팀 이름순 -> 연도순)
final_result = final_result.sort_values(by=['Team', 'Year'])

# 5. CSV 파일 저장 (한글 깨짐 방지 인코딩 적용)
final_result.to_csv('data/KBO_Team_Yearly_Summary.csv', index=False, encoding='utf-8-sig')
```
